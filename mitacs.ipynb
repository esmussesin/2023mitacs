```python
import sympy as sp
x,y = sp.symbols('x y')

eq = x**2 + 2*x + y**2 -4

sol = sp.solve(eq,y)
center = sp.solve(eq,x)
radius = sp.sqrt((center[0]-(center[0]+center[1])/2)**2+sol[0]**2)
print(radius)
print(sol)
print(center)
```

```python
from sympy import Sum,symbols,init_printing,oo,simplify,limit   #we import the certain functions needed from sympy module
def alternating_series():                      #define a function which detect whether the series is converged
  init_printing()                          #to use unicode char
  n = symbols('n')                        #make n as a variable
  S = Sum((-1)**(n+1)/n**2,(n,1,oo))               #sum the series whose general item is (-1)**(n+1)/n**2
  S = simplify(S)                         #simplify the sum-S
  if S.is_finite:                         #Determine whether S is infinite
    print('the series is diverged')
  else:
    print('the series is converged')
  lim = limit(S,n,oo)                       #find the limit of S
  lim = simplify(lim)                       #simplify the limit
  print(lim)
if __name__ == '__main__':
  alternating_series()                       #use main function to call for alternating_series()
```

`the code uses sympy module to determine whether a certain series is converged and print the limit if converged.`

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
t = np.linspace(-10, 10, 100)
x = t*(np.sin(6*t))
y = t*(np.cos(6*t))
z = t
fig = plt.figure()
ax = fig.add_subplot(1,3,(1,3), projection='3d')
ax.plot(x, y, z)
r = np.sqrt(x**2 + y**2)
ax.plot(x, y, r, 'r')
plt.show()
```
